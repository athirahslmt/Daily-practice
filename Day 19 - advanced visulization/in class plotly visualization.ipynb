```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.set(style='darkgrid')
sns.pairplot(tips,hue='time')
```

```python
plt.scatter(x= tips ['total_bill'], y=tips['tip'])
```

```python
sns.set(style='darkgrid')
sns.pairplot(tips, hue='time')
```

```python
sns.pairplot(tips, hue='time')

#rc parameters
#runtime configurations
plt.rcParams['figure.figsize']=( 10,7)
plt.rcParams['lines.linewidth'] = 3
plt.rcParams['lines.linestyle'] = '--'
plt.rcParams['grid.linestyle'] = '--'
```

```python
#join plot

sns.jointplot(tips, x ='total_bill',y='tip',kind='reg',color='orange')

#distribution on top and left of the graph
```

```python
sns.countplot(tips)
```

```python
#plotly
import plotly.express as px

px.bar(tips, x='day', y='total_bill')
```

```python
barplot = px.bar(tips, x='day', y='total_bill')
barplot.show()
```

```python
#group by any of the categorical column

bill_by_day = tips.groupby('day')['total_bill'].mean()
bill_by_day
```

```python
barplot= px.bar(bill_by_day, x='day',y='total_bill')

barplot.show()
```

```python
bill_by_sex = tips.groupby('sex')['total_bill'].max()
bill_by_sex
```

```python
scatterbilltip = px.scatter(
    tips,
    x = 'total_bill',
    y = 'tip',
    size = 'size',
    color = 'day',
    hover_data = ['sex','smoker'],
    facet_col='sex'
)
```
