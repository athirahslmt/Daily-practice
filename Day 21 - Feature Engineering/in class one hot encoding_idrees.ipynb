### One hot encoding

```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
iris = sns.load_dataset('iris')
iris.sample(10)
```

```python
# get dummies method - categories --> Columns
# automatically detects the catagorical columns as well 
df_onehot = pd.get_dummies(iris, columns=['species'])

df_onehot['species_setosa'].dtype

df_onehot.sample(10)
```

Get Dummies, it automatically ignores any missing values

df

```python
df = pd.DataFrame([ 1,2,np.nan,1,1,1,3,3,3,2,2,2,3,3,3,1,1,1,1], columns= ["first"])
df.head()

df['first'].dtype

df['first'] = df['first'].astype('category')
df['first'].dtype

df_encoded = pd.get_dummies(df, columns=['first'], dummy_na=False)
df_encoded.head()
# df_encoded.info()
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

# Creating the encoder
encoder = OneHotEncoder( sparse_output=False, handle_unknown='ignore')

# Ignore the missing values if handle_unknown = 'ignore'"in class label encoding .ipynb"
```

```python
iris_onehot = iris.copy()

encoder = OneHotEncoder( sparse_output = False, handle_unknown='ignore')

catagorical_cols = ['species']

ecoded_values = encoder.fit_transform(iris_onehot[catagorical_cols]) # species

# names of new columns, that don't exist yet
new_columns = encoder.get_feature_names_out(catagorical_cols)
new_columns
# ['species_setosa', 'species_versicolor', 'species_virginica']

# iris_onehot

# adding the columns to the dataframe

iris_onehot[new_columns] = ecoded_values

iris_onehot.drop(columns=['species'], inplace = True)

iris_onehot
```
