### scaling 

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
data = pd.DataFrame({
    'age': [20, 50, 100],
    'income': [20000, 50000, 100000]
})
# upscale/downscale
scaler = MinMaxScaler()
# Normalization 
# you want to check what the data will look like in a scale of 0 to 1.
normalize_data = scaler.fit_transform(data)

df_normalized = pd.DataFrame(normalize_data, columns= ['age', 'income'])
df_normalized
```

- Why do we normalize data?

    - easier to calculate
    - 

- a lot prefernce to salary over age in the given example
- our models only use numbers
