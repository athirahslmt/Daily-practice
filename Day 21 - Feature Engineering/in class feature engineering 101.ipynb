```python
#label encoding
#column that you encode

penguins['species'].unique()
```

```python
# {}

special_codes = {
    'Adelie': 0,
    'Chinstrap' :1,
    'Gentoo' :2
}
```

```python
penguins['species_encoded'] = penguins['species'].map(special_codes)
```
