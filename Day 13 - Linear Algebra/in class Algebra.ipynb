```python
x = np.array([[1,4,1],
             [2,5,2]])
w= np.array([10,10,10])
w2 = np.array([[10,2,3],
             [10,2,3],
             [10,2,3]])

matmultiply_xw=np.dot(x,w)
print(matmultiply_xw)
```

```python
matmultiply_xw2=np.dot(x,w2)
print(matmultiply_xw2)
```

```python
matmultiply_xw3= x@w2
print(matmultiply_xw3)
```

```python
matmultiply_xw4=np.matmul(x,w2)
print(matmultiply_xw4)
```
