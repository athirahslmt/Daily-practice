```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

#Load dataset
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)

#Apply K-Means
kmeans = KMeans(n_clusters=3, init='k-means++',random_state=42)
df['Cluster'] = kmeans.fit_predict(df)

#Visualize cluster

plt.scatter(df.iloc[:,0], df.iloc[:,1], c=df['Cluster'], cmap='viridis')
plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[1])
plt.title('K-Means Clustering on Iris Dataset')
plt.show()
```

```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
#load the iris dataset
iris = load_iris()
x = iris.data #features
y = iris.target
```

```python
#perform hierarchical clustering using ward's method

Z = linkage(x,method='ward')
```

```python
#create dendogram
plt.figure(figsize=(10,7))
dendrogram(Z, labels=iris.target_names[y], leaf_rotation=90, leaf_font_size=12,color_threshold=10)
plt.title('Hierarchical CLustering Dendogram of Iris Dataset')
plt.xlabel('Iris SPecies')
plt.ylabel('Distance')
plt.axhline(y=7, color='r', linestyle='--')#optional line to indicate cutoff
plt.show()
```
